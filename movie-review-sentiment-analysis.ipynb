```python
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv("../input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv")
```

```python
df.dropna()
```

```python
df['review'][0]
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(oov_token="dilip")
```

```python
tokenizer.fit_on_texts(df['review'])
```

```python
tokenizer.word_index
```

```python
tokenizer.document_count
```

```python
# no of unique words
len(tokenizer.word_index)
```

```python
sequences=tokenizer.texts_to_sequences(df['review'])
```

```python
sequences[0]
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
sequences=pad_sequences(
    sequences,
    maxlen=100,
    padding='post',
    truncating='post'
)
```

```python
sequences[0]
```

```python
type(sequences)
```

```python
len(sequences)
```

```python
label=[]
for x in df['sentiment']:
    label.append(1 if x=="positive" else 0)
```

```python
label[:5]
```

```python
label=np.array(label)
```

```python
type(label)
```

```python
label
```

```python
from keras.layers import Dense,SimpleRNN,Embedding,Flatten,Dropout
```

```python
tf.random.set_seed(42)

model=tf.keras.Sequential()

model.add(Embedding(10000, 2,input_length=100))
model.add(SimpleRNN(64,return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(sequences[:10000], label[:10000],epochs=10,validation_data=(sequences[20000:21000],label[20000:21000]))
```

```python
model.evaluate(sequences[49000:],label[49000:])
```
